```python
import pandas as pd
import pickle

# Load metadata
with open('GSE42861_metadata_2023.pkl', 'rb') as f:
    metadata = pickle.load(f)

# Load data
with open('GSE42861.pkl', 'rb') as f:
    data = pickle.load(f)

# Convert to DataFrame
metadata_df = pd.DataFrame(metadata)
data_df = pd.DataFrame(data)


metadata_df
```

```python
# Check the first few rows of the metadata
print(metadata_df.head())

# Check the first few rows of the data
print(data_df.head())

# Check the shape of the data
print("Metadata Shape:", metadata_df.shape)
print("Data Shape:", data_df.shape)

# Summary statistics
print(data_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Disease Status
sns.countplot(data=metadata_df, x='Disease State')
plt.title('Distribution of Disease Status')
plt.show()

# Age distribution
sns.histplot(metadata_df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
X = data_df.filter(regex='^cg', axis=1)
y_disease = metadata_df['Disease State']
y_age = metadata_df['Age']

# Initialize Random Forest Classifier
rf_disease = RandomForestClassifier()
rf_age = RandomForestRegressor()  # For age prediction, using regression

# Fit the model for Disease Status
rf_disease.fit(X, y_disease)

# Fit the model for Age
rf_age.fit(X, y_age)

# Feature importance for Disease Status
feature_importance_disease = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_disease.feature_importances_
})

# Sort features by importance
feature_importance_disease = feature_importance_disease.sort_values(by='Importance', ascending=False)

# Top 10 significant features for Disease Status
top_features_disease = feature_importance_disease.head(10)['Feature'].tolist()
print("Top 10 significant features for Disease Status:", top_features_disease)

# Feature importance for Age
feature_importance_age = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_age.feature_importances_
})

# Sort features by importance
feature_importance_age = feature_importance_age.sort_values(by='Importance', ascending=False)

# Top 10 significant features for Age
top_features_age = feature_importance_age.head(10)['Feature'].tolist()
print("Top 10 significant features for Age:", top_features_age)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Plotting the explained variance ratio
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio.cumsum(), marker='o', linestyle='--')
plt.title('Explained Variance Ratio')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error

# Split data into train and test sets
X_train, X_test, y_train_disease, y_test_disease = train_test_split(X[top_features_disease], y_disease, test_size=0.2, random_state=42)
_, _, y_train_age, y_test_age = train_test_split(X[top_features_age], y_age, test_size=0.2, random_state=42)

# RandomForest for Disease Status
rf_model_disease = RandomForestClassifier()
rf_model_disease.fit(X_train, y_train_disease)
rf_pred_disease = rf_model_disease.predict(X_test)
print("RandomForest Accuracy for Disease Status:", accuracy_score(y_test_disease, rf_pred_disease))

# GradientBoosting for Age
gb_model_age = GradientBoostingRegressor()
gb_model_age.fit(X_train, y_train_age)
gb_pred_age = gb_model_age.predict(X_test)
mse = mean_squared_error(y_test_age, gb_pred_age)
print("GradientBoosting MSE for Age:", mse)
```
